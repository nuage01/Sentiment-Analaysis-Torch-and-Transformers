```python
from flask import Flask
from flask_sqlalchemy import SQLAlchemy

app = Flask(__name__)
app.config["SQLALCHEMY_DATABASE_URI"] = "sqlite:///bdd/sql_database/ml_processed.db"
db = SQLAlchemy(app)


```

```python
db
```

```python

db.Model.metadata.reflect(db.engine)
```

```python
class Reviews(db.Model):
    __tablename__ = 'reviews'
    __table_args__ = { 'extend_existing': True }  
```

```python
Reviews.query.all()
users = Reviews.query.all()

for user in users:
    print (user.topic)
```

```python
from functools import reduce
targets = ['ALLERGENS', 'COMPETITION', 'DELIVERY', 'PACKAGING', 'PRICE', 'TASTE']
for user in users:
    if user.topic:
        user.topic = reduce(lambda x,y: x + ', ' + y, [targets[int(topic)-1] for topic in user.topic.split(',') ])
    else: 
        user.topic = "No topic related"
```
